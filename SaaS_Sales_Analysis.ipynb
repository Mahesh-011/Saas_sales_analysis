# Saas Sales Analysis.

```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['Month-Year'] = df['Order Date'].dt.to_period('M')
df
```

```python
df.drop(['Row ID', 'Order ID', 'Date Key', 'Customer ID'], axis=1, inplace=True)
df
```

```python
print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df.info()
df.head()
```

```python
#Sales Trend Over Time

monthly_sales = df.groupby('Month-Year')['Sales'].sum().reset_index()
monthly_sales['Month-Year'] = monthly_sales['Month-Year'].astype(str)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month-Year', y='Sales', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.tight_layout()
plt.show()
```

```python
#Top Selling Products

top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(10, 5), title='Top 10 Products by Sales')
plt.ylabel("Total Sales")
plt.xlabel("Product")
plt.tight_layout()
plt.show()
```

```python
# Most Profitable Customers
top_customers = df.groupby('Customer')['Profit'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar', figsize=(10, 5), title='Top 10 Customers by Profit')
plt.ylabel("Profit")
plt.xlabel("Customer")
plt.tight_layout()
plt.show()
```

```python
#Sales By Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar', figsize=(8, 5), title='Sales by Region')
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
#Segment wise Revenue Breakdown
segment_sales = df.groupby('Segment')['Sales'].sum()
segment_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(5, 5), title='Revenue Share by Segment')
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
# Top Performing Products by Region
top_products = df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()
top_products = top_products.sort_values(by='Sales', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Sales', hue='Product', data=top_products.head(10))
plt.title("Top 10 Products by Sales across Regions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Which Countries are Most Profitable.

country_profit = df.groupby('Country')['Profit'].sum().sort_values(ascending=False).head(10)
country_profit.plot(kind='bar', figsize=(10, 5), color='orange')
plt.title('Top 10 Countries by Total Profit')
plt.ylabel("Profit")
plt.xlabel("Country")
plt.tight_layout()
plt.show()
```

```python
#Which Segments & Industries Generate the Most Profit?
seg_ind_profit = df.groupby(['Segment', 'Industry'])['Profit'].sum().unstack().fillna(0)
seg_ind_profit.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Profit by Segment and Industry')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Profitability by License Type
license_profit = df.groupby('License')['Profit'].sum().sort_values(ascending=False).head(10)
license_profit.plot(kind='bar', figsize=(10, 5), color='teal')
plt.title('Top Licenses by Profit')
plt.ylabel("Profit")
plt.xlabel("License")
plt.tight_layout()
plt.show()
```

```python
#Discount vs Profit Relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.tight_layout()
plt.show()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
top_country = df.groupby('Country')['Profit'].sum().idxmax()
top_product = df.groupby('Product')['Sales'].sum().idxmax()

summary = {
    'Total Sales': [round(total_sales, 2)],
    'Total Profit': [round(total_profit, 2)],
    'Top Country by Profit': [top_country],
    'Top Product by Sales': [top_product]
}

pd.DataFrame(summary)
```
